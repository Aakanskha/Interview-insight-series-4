# Interview Insight Series-4

# ** # Question 1: Microsoft Teams – Reduce Meeting Fatigue for Remote Workers**
You’re a Product Manager at Microsoft working on Teams. After the pandemic, internal research shows growing “meeting fatigue” among remote employees—leading to reduced engagement and higher attrition in large enterprises. Leadership asks you to define a product initiative to tackle this Draft the PRD, define user journeys and JTBD, and propose success metrics and OKRs

  Ans) Problem Statement: Remote workers are experiencing "meeting fatigue," leading to cognitive overload, decreased productivity, and higher attrition.

1. Jobs-to-be-Done (JTBD)
 - "When I have a back-to-back schedule, I want to capture key decisions without attending every minute, so that I can focus on deep work without missing out."

 - "When I am in a long meeting, I want interactive elements that keep me focused, so that I don't feel like a passive observer."

2. User Journeys

 - Pre-Meeting: The system flags "High-Density Days" and suggests shortening 30-min meetings to 25-min to allow for "bio-breaks."
 - In-Meeting: AI-powered "Focus Mode" silences non-urgent notifications and provides a real-time "Energy Meter" to the organizer to suggest breaks.
 - Post-Meeting: A "Catch-up Dashboard" provides a 2-minute AI summary and specific @mentions, allowing users to skip the recording.

 3. Product Requirements Document (PRD) Snippet

  - Feature Name: Teams "Smart Decoupling" & Wellness Integration.
  - Goal: Reduce synchronous meeting time by 15% through asynchronous summaries.
  - Target Audience: Corporate employees in large enterprises.
  - Functional Specs: * AI Recap: Auto-generation of action items sent via chat.
  -> Meeting Privacy: "Quiet Hours" enforcement that auto-declines meetings during deep-work blocks.

  4. Success Metrics & OKRs

  -> Objective: Improve employee well-being and engagement via Microsoft Teams.
   - KR 1: Increase "Deep Work" hours (unscheduled blocks) by 20% per user per week.
   - KR 2: Reduce the average duration of internal meetings from 45 mins to 35 mins.
   - North Star Metric: Meeting-to-Task Completion Ratio (Higher efficiency with fewer hours spent).

# ** # Question 2: Adobe Express – Empower Creators with Zero Design Skills**
You are a PM at Adobe Express (a Canva competitor). Data shows that small business owners and solopreneurs find Adobe tools “intimidating” and often abandon after onboarding. Your job is to design a self-guided, lowfriction creation experience to retain these users. Define your PRD, OKRs, and how you would discover the right solutions and JTBD.

Ans) Problem Statement: Small business owners abandon the platform due to the high "intimidation factor" of the professional-grade UI.

1) Solution Discovery & JTBD
To discover solutions, I would use Gen-AI Prompt Analysis (seeing what users try to describe but can't build) and Churn Surveys.
 - JTBD: "When I need a social media post for my sale, I want to describe my idea in plain English and get a 90% finished design, so I don't have to learn layers and typography."

2) Product Requirements Document (PRD) Snippet

-> Feature Name: "Express Magic-Start" Onboarding.
-> Goal: Transition from a "Blank Canvas" to a "Guided Prompt" experience.
-> Key Features:
 - Industry-Specific Onboarding: Upon login, ask: "What do you sell?" (e.g., Coffee, Real Estate).
 - One-Tap Brand Kit: Auto-extract colors and logos from the user’s URL to eliminate setup friction.
 - Contextual Tooltips: Hide advanced features (like 'Kerning') under an "Advanced" toggle to keep the primary UI clean.

3) Success Metrics & OKRs
-> Objective: Become the go-to design tool for non-designers by reducing "Time-to-Value."
 - KR 1: Increase Day-1 Retention from $X\%$ to $Y\%$ by shortening the first-design completion time to under 3 minutes.
 - KR 2: Achieve a 30% increase in "Successful Exports" (Designs saved/shared) during the first session.
 - KR 3: Reduce the average number of clicks from "Landing" to "First Edit" by 50%.

4) Navigation Strategy (The Low-Friction Path)
 - The "Guided" vs. "Manual" Toggle: Allow users to choose their path at the start.
 - AI-Copilot: A persistent chat-based design assistant that handles complex tasks (e.g., "Remove the background and make it look like it's on a beach") via voice or text commands.
